###5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

####a.Ploting classification based on SL


```python
plt.scatter(df['Classification'],df['SL'])
plt.title('On the basis of SL' ,fontsize= 15)
plt.xlabel('Classification',fontsize =12)
plt.ylabel('SL', fontsize = 12)
```

iris-virginica has the largest SL and iris-setosa has the lowest.

###b.Classification of the three species using SL and SW

```python
sns.scatterplot(x = 'SW', y = 'SL', hue = 'Classification', data = df,
                    palette = ['b', 'r', 'g'])
plt.title('CLASSIFICATION on SW & SL')
```

iris-setosa has the highest SW and lowest SL,
iris-versicolor has almost average value for both SW and SL,
iris-virginica is having the heighest SL.

### c. Ploting classification based on PL

```python
plt.scatter(df['Classification'],df['PL'],c='brown',marker='^')
plt.title('On the basis of PL' ,fontsize= 18,c='brown')
plt.xlabel('Classification',fontsize =15,c='brown')
plt.xticks(fontsize=10)
plt.ylabel('PL', fontsize = 15,c='brown')
```

iris-virginica has the largest PL while iris-setosa has the lowest.

### d.Corelation between SW,SL,PL,PW


```python
df.corr()
```

```python
col=df[['SL','SW','PL','PW']]
sns.heatmap(col.corr(), cmap= 'coolwarm',annot = True)
```

Higher intensity of the color red represents higher corelation while the color blue represents negative corelation and the lighter color represnts slight correation.


###e. Classification on account of PW

```python
plt.scatter(df['Classification'],df['PW'],c='brown',marker='^')
plt.title('On the basis of PW' ,fontsize= 18,c='brown')
plt.xlabel('Classification',fontsize =15,c='brown')
plt.xticks(fontsize=10)
plt.ylabel('PW', fontsize = 15,c='brown')
```

iris-virginica: highest,
iris-versicolor: average,
iris-setosa: lowest
