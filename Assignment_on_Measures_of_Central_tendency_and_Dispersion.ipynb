3. Who scores the most on average for math, reading and writing based on
● Gender
● Test preparation course

```python
# scores based on gender
sp_gender = sp[['gender','math score','reading score','writing score']].groupby('gender').mean()
sp_gender
```

So while male students are having most on average for math scores, female students are having the most on average for the remaing test scores ie. for reading and writing tests.

```python
#scores based on test preparation course
sp_tpc = sp[['test preparation course','math score','reading score','writing score']].groupby('test preparation course').mean()
sp_tpc
```

Based on the test preparation course, those students who have completed the course have scored most on average on all the tests.

4. What do you think about the scoring variation for math, reading and writing
based on
● Gender
● Test preparation course

```python
# scoring variation based on gender
sp_genderV = sp[['gender','math score','reading score','writing score']].groupby('gender').std()
sp_genderV
```

```python
# Plotting
sp_genderV.plot(kind='bar')
plt.title('Standard Deviation of Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Standard Deviation')
plt.legend(title='Subject')
```

The scoring variation of mark scores for male students dosen't seems much but for female students the math score test deviates more from the other two scores.

```python
#scoring variation based on test preparation course
sp_tpc = sp[['test preparation course','math score','reading score','writing score']].groupby('test preparation course').std()
sp_tpc
```

```python
#plotting
sp_tpc.plot(kind='bar')
plt.title('Standard Deviation of Scores by Test Preparation Course')
plt.xlabel('Test Preparation Course')
plt.ylabel('Standard Deviation')
plt.legend(title='Subject')
```

The variance of scores of 3 tests among those who completed and not attempted the test preparion course doesn't differ much from each other.
